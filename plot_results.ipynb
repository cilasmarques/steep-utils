## Aux functions

```python
def plot_per_execution(df, title, y_title, x_title, col):
    fig = go.Figure()
    
    for phase in df['PHASE'].unique():
        fig.add_trace(go.Box(
            y=df[df['PHASE'] == phase][col],
            x=df[df['PHASE'] == phase]['threads'],
            name=phase
        ))

    fig.update_layout(
        width=1500,
        title=title,
        yaxis_title=y_title,
        xaxis_title=x_title,
        boxmode='group' # group together boxes of the different traces for each value of x
    )

    fig.show()
```

```python
def plot_per_phase(df, title, y_title, x_title, col):
    fig = go.Figure()

    for thread in df['threads'].unique():
        fig.add_trace(go.Box(
            y=df[df['threads'] == thread][col],
            x=df[df['threads'] == thread]['PHASE'],
            name=thread
        ))

    fig.update_layout(
        width=1500,
        title=title,
        yaxis_title=y_title,
        xaxis_title=x_title,
        boxmode='group' # group together boxes of the different traces for each value of x
    )

    fig.update_xaxes(type='category', categoryorder='array', categoryarray=['P1 - TOTAL', 'P2 - PIXEL SELECTION', 'P2 - RAH CYCLE', 'P2 - H', 
                                                                         'P2 - FINAL PRODUCTS', 'P2 - TOTAL', 'TOTAL'])

    fig.show()
```

## Plots

#### Timestamp

```python
plot_per_phase(dfs, "Phases timestamp", "Time (ms)", "phase", "TIMESTAMP")
```

```python
plot_per_execution(dfs, "Execution timestamps", "Time (ms)", "phase", "TIMESTAMP")
```

#### GPU

```python
plot_per_execution(dfs, "Execution timestamps", "Time (ms)", "phase", "GPU_MAX")
```
